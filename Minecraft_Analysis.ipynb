````python
# Etapa 1: Importar bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

# Etapa 2: Criar os dados simulados
df_growth = pd.DataFrame({
    "year": [2019, 2020, 2021, 2022, 2023, 2024],
    "active_players_millions": [91, 126, 141, 170, 180, 185]
})

df_ages = pd.DataFrame({
    "age_group": ["<15", "15-21", "22-30", "31+"],
    "percentage": [19, 39, 25, 7]
})

df_platforms = pd.DataFrame({
    "platform": ["Mobile", "PC", "Console"],
    "percentage": [40, 35, 25]
})

df_season = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
              "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "average_players_millions": [145, 146, 148, 150, 152, 153, 180, 179, 160, 158, 162, 175]
})
````

````python
# Etapa 3: Visualizações
sns.lineplot(data=df_growth, x="year", y="active_players_millions", marker="o")
plt.title("📈 Jogadores Ativos por Ano")
plt.xlabel("Ano")
plt.ylabel("Milhões de Jogadores")
plt.show()
````

````python
sns.barplot(data=df_ages, x="age_group", y="percentage", palette="Blues_d")
plt.title("👶👦👨‍🦰 Distribuição Etária dos Jogadores")
plt.xlabel("Faixa Etária")
plt.ylabel("Percentual (%)")
plt.show()
````

````python
sns.barplot(data=df_platforms, x="platform", y="percentage", palette="Set2")
plt.title("🎮 Dispositivos mais usados para jogar Minecraft")
plt.ylabel("Percentual (%)")
plt.xlabel("Plataforma")
plt.show()
````

````python
sns.lineplot(data=df_season, x="month", y="average_players_millions", marker="o", color="green")
plt.title("📅 Atividade Mensal dos Jogadores (2024)")
plt.ylabel("Milhões de Jogadores")
plt.xlabel("Mês")
plt.show()
````

## 📌 Conclusão

O Minecraft continua sendo um dos jogos mais populares globalmente, com uma base crescente e bem distribuída em idade e plataformas.

As análises mostram:

- Crescimento estável de jogadores desde 2019
- Predominância de jogadores jovens (15 a 21 anos)
- Alta adesão via mobile
- Sazonalidade clara nas férias escolares
